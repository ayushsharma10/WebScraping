```python
#Libraries
import os
import pandas as pd
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
#Initialising

url = 'https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_suicide_rate'
uClient = uReq(url)
page_html = uClient.read()
uClient.close()

page_soup = soup(page_html,'html.parser')
table = page_soup.findAll('table', {'class':'wikitable sortable'})[0]

#Headers
headers = []
for title in table.tr.findAll('th'):
    headers.append(title.text.strip())
headers
```

```python
# Extracting data for each row
rank = []
state = []
suicide = []

for row in table.findAll('tr')[1:]:
    cell = row.findAll('td')
    rank.append(cell[0].text)
    state.append(cell[1].text)
    suicide.append(float(cell[2].text.strip()))

    
```

```python
#Reshaping and formating data
data = pd.DataFrame([rank, state, suicide]).T
data.columns = headers
```

```python
#Let's have a look at the results
data
```

```python
#Saving data

data.to_csv('sucides.csv', index=False)
```
